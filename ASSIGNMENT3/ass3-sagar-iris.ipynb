```python
import pandas as pd
import numpy as np

file_path = '/home/csl-4/Documents/7348/ASSIGNMENT3/Iris.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
# Group by Species and calculate summary statistics
summary_stats = df.groupby('Species')[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].agg(['mean','median','min','max','std'])
print("Summary Statistics by Species:\n", summary_stats)
```

```python
numeric_lists = {sp: df[df['Species']==sp][['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values.tolist() for sp in df['Species'].unique()}
print("\nNumeric Lists by Species:\n", numeric_lists)
```

```python
for sp in df['Species'].unique():
    petal = df[df['Species']==sp]['PetalLengthCm']
    print(f"\nSpecies: {sp}")
    print("Mean:", round(petal.mean(), 2))
    print("Std Dev:", round(petal.std(), 2))
    print("25th percentile:", np.percentile(petal, 25))
    print("Median (50th percentile):", np.percentile(petal, 50))
    print("75th percentile:", np.percentile(petal, 75))
```
